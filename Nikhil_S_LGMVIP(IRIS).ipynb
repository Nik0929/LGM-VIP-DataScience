# Iris Dataset Classfication

#### Problem statement : To classfiy iris plant accroding to sepal and petal length

step 1 : importing required libraries

```python
import numpy as np
import pandas as pd
```

step 2: import dataset

```python
names=['sepal-length','sepal-width','petal-length',
       'petal-width','Class']
data=pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=names)

data
```

```python
x=data.iloc[:,:4].values
y=data.iloc[:,4:5]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
data['Class'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
